```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
housing_df = fetch_california_housing()
housing_df
```

```python
housing_df.keys()
```

```python
print(housing_df.DESCR)
```

```python
housing_df["data"]
```

### Input Data

```python
housing_df["feature_names"]
```

### Output Data (Price Prediction)

```python
housing_df["target_names"]
```

```python
df = pd.DataFrame(housing_df["data"], columns=housing_df["feature_names"])
df
```

```python
df["Price"] = housing_df["target"]
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
sns.heatmap(df.corr(), annot=True)
plt.show()
```

```python
x = df.drop(columns="Price")           # x = df.iloc[:, :-1]   or   x = df[df.columns[:-1]]
y = df[["Price"]]                      # y = df.iloc[:, -1]    or   y = df[df.column[-1]]
display(x.head(),y.head())
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
sns.scatterplot(x_train)
plt.title("x_train", color="red")

plt.subplot(1,2,2)
sns.scatterplot(x_train_scaled)
plt.title("x_train after Scaling", color= "red")
plt.show()
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error, r2_score

linear_reg = LinearRegression()
linear_reg.fit(x_train_scaled, y_train)
y_pred = linear_reg.predict(x_test_scaled)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = root_mean_squared_error(y_test, y_pred)

score = r2_score(y_test, y_pred)

print(f"Mean Absolute Error : {mae} \nMean Squared Error : {mse} \nRoot Mean Squared Error : {rmse} \n\nR2 Score : {score}")
```

```python
plt.scatter(y_test, y_pred, color="cyan")
plt.show()
```

```python
residuals = y_test-y_pred
residuals
```

```python
sns.displot(residuals, kind="kde")
plt.show()
```

* From the Above graph we can say that our model is performing well.

# Pickling
* Python pickle module is used for serializing and de-serializing a python object structure. Any object in python can be pickled so that it can be saved on local disk. What pickle does is that it "Serialises" the object first before writing it to file. Pickling is a way to convert a python object (list, dict, etc.) into a character stream. The idea is that this character stream contains all the information necessary to reconstruct the object in another python script.

## Saving Pickle File in Local Disk

```python
import pickle
pickle.dump(linear_reg, open("House_Price_Prediction_Model.pkl", "wb"))
```

## Loading Pickle File

```python
model = pickle.load(open("House_Price_Prediction_Model.pkl", "rb"))
```

## Predicting House Price

```python
model.predict(x_test)
```
