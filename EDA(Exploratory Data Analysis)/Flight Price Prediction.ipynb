### Drop "Route" Column.

```python
df.drop("Route", axis=1, inplace=True)
df
```

### Split Date of Journey

```python
df["Date"] = df["Date_of_Journey"].str.split("/").str[0]
df["Month"] = df["Date_of_Journey"].str.split("/").str[1]
df["Year"] = df["Date_of_Journey"].str.split("/").str[2]
df
```

```python
df.info()
```

### We need to convert "Date", "Month", "Year" to int.

```python
df["Date"] = df["Date"].astype(int)
df["Month"] = df["Month"].astype(int)
df["Year"] = df["Year"].astype(int)
df.info()
```

### Drop "Date of Journey" Column.

```python
df.drop("Date_of_Journey", inplace=True, axis=1)
df
```

### Split Departure Time

```python
df["Dept_Hr"] = df["Dep_Time"].str.split(":").str[0]
df["Dept_Min"] = df["Dep_Time"].str.split(":").str[1]
df
```

### Convert "Dept_Hr" and "Dept_Min" to int.

```python
df["Dept_Hr"] = df["Dept_Hr"].astype(int)
df["Dept_Min"] = df["Dept_Min"].astype(int)
df.info()
```

### Drop "Dept_Time" Column.

```python
df.drop("Dep_Time", axis=1, inplace=True)
df
```

### Removing unwanted data and then splitting the "Arrival_Time" column.

```python
df["Arr_Hr"] = df["Arrival_Time"].str.split(" ").str[0].str.split(":").str[0]
df["Arr_Min"] = df["Arrival_Time"].str.split(" ").str[0].str.split(":").str[1]
df
```

### Converting "Arr_Hr" and "Arr_Min" into int.

```python
df["Arr_Hr"] = df["Arr_Hr"].astype(int)
df["Arr_Min"] = df["Arr_Min"].astype(int)
df.info()
```

### Drop "Arrival_Time" Column.

```python
df.drop("Arrival_Time", axis=1, inplace=True)
df
```

### Convert "Duration" Column to "Duration(Minutes)"

```python
df["Duration_Hr"] = df["Duration"].str.split(" ").str[0].str[:-1]
df["Duration_Min"] = df["Duration"].str.split(" ").str[1].str[:-1]
df
```

### Fill Nan values with 0 in "Duration_Min" Column.

```python
df.fillna({"Duration_Min" : 0}, inplace=True)
df
```

### Convert "Duration_Hr" and "Duration_Min" to int.

```python
df["Duration_Hr"] = df["Duration_Hr"].astype(int)
df["Duration_Min"] = df["Duration_Min"].astype(int)
df.info()
```

### Create new column with "Duration_in_Min".

```python
df["Duration_in_Min"] = (df["Duration_Hr"]*60)+df["Duration_Min"]
df
```

### Drop "Duartion", "Duration_Hr" and "Duration_Min" Columns.

```python
df.drop(["Duration", "Duration_Hr", "Duration_Min"], axis=1, inplace=True)
df
```

### "Total_Stops" Column.

```python
df[df["Total_Stops"].isna()]
```

```python
df["Total_Stops"].unique()
```

```python
df["Total_Stops"].mode()
```

### Using Map Function and replacing "Nan" Values with mode Values.

```python
df["Total_Stops"] = df["Total_Stops"].map({"non-stop" : 0, "1 stop" : 1, "2 stops" : 2, "3 stops" : 3, "4 stops" : 4, np.nan : 1})
df
```

### Converting "Total_Stops" column to int.

```python
df["Total_Stops"] = df["Total_Stops"].astype(int)
df.info()
```

### Converting Categorical data into Numerical data using OneHotEncoder.

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder()
```

```python
df1 = pd.DataFrame(encoder.fit_transform(df[["Airline", "Source", "Destination"]]).toarray(), columns=[encoder.get_feature_names_out()])
df1
```

### Merge Categorical Data Frame and Previous Data Frame.

```python
res_df = pd.concat([df1,df], axis=1)
res_df
```

### Now Drop the Categorical Data Columns. i.e "Airline", "Source", "Destination".
* This Final Data Frame is ready for Meachine Learning Training.
* Beacuse the Data frame only contains Numerical Data.

```python
res_df.drop(["Airline", "Source", "Destination"], axis=1, inplace=True)
res_df
```
