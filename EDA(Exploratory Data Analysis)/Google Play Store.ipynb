```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Krishnaik06/playstore-Dataset/main/googleplaystore.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

### There is a string in "Reviews" Column.

```python
df[~df["Reviews"].str.isnumeric()]
```

```python
df.drop(df.index[10472], inplace=True)
df.shape
```

```python
df[~df["Reviews"].str.isnumeric()]
```

### Convert "Reviews" Column to int.

```python
df["Reviews"] = df["Reviews"].astype(int)
df.info()
```

```python
df.head()
```

### Check "Size" Column

```python
df["Size"].str[-1].unique()
```

```python
df[df["Size"].str[-1] == "M"]
```

```python
df["Size"] = df["Size"].str.replace("M", "000")
df["Size"] = df["Size"].str.replace("k", "")
df["Size"] = df["Size"].str.replace("+", "")
df["Size"].str[-1].unique()
```

```python
df[df["Size"].str[-1] == "e"]
```

```python
df["Size"] = df["Size"].replace("Varies with device", np.nan)
df["Size"].str[-1].unique()
```

```python
df["Size"] = df["Size"].str.replace(",", "")
```

```python
df.rename(columns={"Size" : "Size(k)"}, inplace=True)
df.head()
```

### Convert "Size" column to float.

```python
df["Size(k)"] = df["Size(k)"].astype(float)
df.info()
```

```python
df.head()
```

### Similarly Check for "Installs" and "Price" Column.

```python
print(df["Installs"].unique(), "\n")
print(df["Price"].unique())
```

```python
df.rename(columns={"Price" : "Price_in_$"}, inplace=True)
```

### Remove Special Characters from "Installs" and "Price" Columns.

```python
spl_char = [",", "+", "$"]
columns = ["Installs", "Price_in_$"]

for i in spl_char:
    for j in columns:
        df[j] = df[j].str.replace(i, "")

print(df["Installs"].unique(), "\n")
print(df["Price_in_$"].unique())
```

```python
df.head()
```

### Convert "Installs" column to int and "Price" column to float.

```python
df["Installs"] = df["Installs"].astype(int)
df["Price_in_$"] = df["Price_in_$"].astype(float)
df.info()
```

### Check "Last Updated" Column

```python
df["Last Updated"].unique()
```

```python
df["Last Updated"] = pd.to_datetime(df["Last Updated"])
df.head()
```

```python
df["Last Updated"] = df["Last Updated"].astype(str)
```

```python
df["Day"] = df["Last Updated"].str.split("-").str[-1]
df["Month"] = df["Last Updated"].str.split("-").str[1]
df["Year"] = df["Last Updated"].str.split("-").str[0]
df.head()
```

```python
df["Day"] = df["Last Updated"].str.split("-").str[-1].str[:2]
df.head()
```

### Convert "Day", "Month", "Year" Columns to int.

```python
df["Day"] = df["Day"].astype(int)
df["Month"] = df["Month"].astype(int)
df["Year"] = df["Year"].astype(int)
df.info()
```

### Remove "Last Updated" Column.

```python
df.drop("Last Updated", axis=1, inplace=True)
df.head()
```

```python
df[df["App"].duplicated()]
```

### Drop Duplicate Records

```python
df["App"].drop_duplicates(inplace=True)
df.shape
```

```python
df.drop_duplicates(inplace=True, subset=["App"])
df.reset_index(inplace=True, drop=True)
df
```

```python
df.to_csv("cleaned_play_store.csv", index=False)
```
