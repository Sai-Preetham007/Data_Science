### Assignment 4: Data Aggregation

1. Create a Pandas DataFrame with 2 columns: 'Category' and 'Value'. Fill the 'Category' column with random categories ('A', 'B', 'C') and the 'Value' column with random integers. Group the DataFrame by 'Category' and compute the sum and mean of 'Value' for each category.
2. Create a Pandas DataFrame with 3 columns: 'Product', 'Category', and 'Sales'. Fill the DataFrame with random data. Group the DataFrame by 'Category' and compute the total sales for each category.

```python
arr = np.array([[random.choice(["A", "B", "C"]) for i in range(10)], [random.randint(1,20) for i in range(10)]])
df = pd.DataFrame(arr)
display(df)

arr = arr.T      # T ---> Transpose.
df = pd.DataFrame(arr, columns=["Category", "Values"])
display(df)


df["Values"] = pd.to_numeric(df["Values"])
x = df.groupby("Category")["Values"].sum()
print(f"Category-wise Sum :")
display(x)

y = df.groupby("Category")["Values"].mean()
print(f"Category-wise Mean :")
display(y)
```

```python
df = pd.DataFrame({"Product" : np.random.choice(["Samsumg", "Apple", "Ferrari", "Lamborgini", "Nike", "Puma"], size=20),
                   "Category" :np.random.choice(["Electronics", "Automobiles", "Footwear"], size=20),
                   "Sales" : np.random.randint(5000, 100000, size=20)}, index=[i for i in range(1,21)])

display(df)

display(df.groupby("Category")["Sales"].sum())
```

### Assignment 5: Merging DataFrames

1. Create two Pandas DataFrames with a common column. Merge the DataFrames using the common column.

```python
df1 = pd.DataFrame({"Product" : np.random.choice(["Samsung", "Apple", "Nike", "Puma", "Ferrari", "Lamborghini"], size=10),
                    "Category" : np.random.choice(["Electronics", "Foot Wear", "Auto Mobiles"], size=10)})
display(df1)

df2 = pd.DataFrame({"Product" : np.random.choice(["Samsung", "Apple", "Puma", "Nike", "Ferrari", "Lamborghini"], size=10),
                    "Sales" : np.random.randint(5000, 50000, size=10)})
display(df2)

concat_rows_df = pd.concat([df1,df2])
display(concat_rows_df)

concat_columns_df = pd.concat([df1,df2], axis=1)
display(concat_columns_df)
```

### Assignment 6: Time Series Analysis

1. Create a Pandas DataFrame with a datetime index and one column filled with random integers. Resample the DataFrame to compute the monthly mean of the values.
2. Create a Pandas DataFrame with a datetime index ranging from '2021-01-01' to '2021-12-31' and one column filled with random integers. Compute the rolling mean with a window of 7 days.

```python
df = pd.DataFrame({"Date" : pd.date_range(start="2022-01-01", end="2022-12-31", freq="D")})
df["Values"] = np.random.randint(0,100, size=len(df))
print("Data Frame :")
display(df)

df1 = df.copy()
df1["Date"] = df["Date"].astype(str)
df1["Month"] = [i.split("-")[1] for i in df1["Date"]]
monthly_mean = df1.groupby("Month")["Values"].mean()
monthly_mean = monthly_mean.reset_index()
monthly_mean.rename(columns={"Values" : "Mean"}, inplace=True)
display(monthly_mean)
```
