```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = 6*np.random.rand(100, 1) - 3
y = 0.5*x**2 + 1.5*x + 2 + np.random.rand(100,1)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

regression = LinearRegression()
```

```python
regression.fit(x_train, y_train)

print(f"Slope or Coefficient : {regression.coef_} \nIntercept or Constant : {regression.intercept_}")
```

```python
y_pred = regression.predict(x_test)
y_pred[:5]
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
print("R2 Score :", r2)
```

```python
plt.scatter(x_train, y_train)
plt.plot(x_train, regression.predict(x_train), color="violet")
plt.show()
```

# Polynomial Features

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2, include_bias=True)
poly
```

```python
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)
```

```python
regression_poly = LinearRegression()
regression_poly.fit(x_train_poly, y_train)
y_pred_poly = regression_poly.predict(x_test_poly)
r2_poly = r2_score(y_test, y_pred_poly)
print("R2 Score :", r2_poly)
```

```python
plt.scatter(x_train, y_train, color="magenta")
plt.scatter(x_train, regression_poly.predict(x_train_poly), color="green")
plt.show()
```
